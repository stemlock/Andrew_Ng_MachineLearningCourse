## 2. Linear Regression with One Variable

In this part of this exercise, you will implement linear regression with one
variable to predict profits for a food truck. Suppose you are the CEO of a
restaurant franchise and are considering different cities for opening a new
outlet. The chain already has trucks in various cities and you have data for
profits and populations from the cities.

The first column is the population of a city and the second column is
the profit of a food truck in that city. A negative value for profit indicates a
loss.

```python
data = np.loadtxt('machine-learning-ex1/ex1/ex1data1.txt', delimiter=",")
X, y = data[:, 0], data[:, 1]
```

### 2.1 Plot the data

```python
plt.scatter(X, y, marker='x', c='red', linewidths=0.75)
plt.ylim(-5, 25)
plt.xlim(4, 24)
plt.title('Figure 1: Scatter plot of training data')
plt.xlabel("Profit in $10,000s")
plt.ylabel("Population of City in 10,000s")
plt.show()
```

### 2.2 Gradient Descent

```python
# Add a column of ones to X as X_0
X_col = np.hstack((np.ones((X.shape[0], 1)), X.reshape(-1,1)))

# Change y into a column vector
y_col = y.reshape(-1,1)

# Initiliaze the parameter vector
theta = np.zeros((1,2))

# Set the number of iterations and alpha learning rate for GD
iters = 1500
alpha = 0.01
```

```python
# Define a cost function
def cost_function(X, y, theta):
    
    # Find the hypothesis function via cross product of X and theta tranposed
    hypothesis = np.matmul(X, theta)
    
    # Find the squared residuals by subtracting actual from predicted and squaring the difference
    sq_residuals = np.power(hypothesis - y, 2)
    
    # Return the sum of the squared residuals divided by 2m
    return np.sum(sq_residuals) / (2*len(y))    
```

```python
cost_function(X_col, y_col, theta.T)
```

```python
# Define the gradient descent function

def gradient_descent(X, y, theta, alpha, iterations):
    
    thetas = [theta]
        
    for i in range(iterations):  
                
        # Find the hypothesis function via cross product of X and theta tranposed
        hypothesis = np.matmul(X, theta)
        
        # Calculate the value of the derivative of the cost function
        cost_derivative = np.sum((hypothesis-y)*X, axis=0)
        
        # Simultaneously update the theta parameter vector
        theta = theta - ((alpha/len(y))*cost_derivative).reshape(-1,1)
        thetas.append(theta)
    
    return thetas
```

```python
thetas = gradient_descent(X_col, y_col, theta.T, alpha, iters)
final_theta = thetas[-1]
print(final_theta)
```

### 2.3 Fitting the Regression Line using the Theta Parameters

```python
line = final_theta[0] + X*final_theta[1]

plt.scatter(X, y, marker='x', c='red', linewidths=0.75, label='Training data')
plt.plot(X, line, '-b', label='Linear Regression line')
plt.ylim(-5, 25)
plt.xlim(4, 24)
plt.title('Figure 1: Training data with linear regression fit')
plt.xlabel("Profit in $10,000s")
plt.ylabel("Population of City in 10,000s")
plt.legend()
plt.show()
```

### 2.4 Visualization the Cost Function

To understand the cost function J($\theta$) better, you will now plot the cost over
a 2-dimensional grid of $\theta_0$ and $\theta_1$ values.

```python
# Create a grid of theta values
theta_0 = np.linspace(-10,10,100)
theta_1 = np.linspace(-1,4,100)

# Initialize the array of cost function values
J_vals = np.zeros((len(theta_0), len(theta_1)))

# For each theta value pair, calculate the cost function
for i in range(len(theta_0)):
    for j in range(len(theta_1)):
        t = np.array([[theta_0[i]], [theta_1[j]]])
        J_vals[i][j] = cost_function(X_col, y_col, t)
        
```

```python
# Plot a surface plot of the cost function

fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(theta_1, theta_0, J_vals, cmap = 'coolwarm', linewidth=0)
fig.show()
```

```python
# Plot a contour graph of the cost function
plt.contour(theta_0, theta_1, J_vals, levels = np.logspace(-2,3,100))
plt.show()
```

```python
# Test a greater number of iterations, we see there is not much difference
thetas = gradient_descent(X_col, y_col, theta.T, 0.01, 1000000)
final_theta = thetas[-1]
print(final_theta)
```

```python
line = final_theta[0] + X_col[:,1]*final_theta[1]

plt.scatter(X, y, marker='x', c='red', linewidths=0.75, label='Training data')
plt.plot(X, line, '-b', label='Linear Regression line')
plt.ylim(-5, 25)
plt.xlim(4, 24)
plt.title('Figure 1: Training data with linear regression fit')
plt.xlabel("Profit in $10,000s")
plt.ylabel("Population of City in 10,000s")
plt.legend()
plt.show()
```

## 3. Linear Regression with Multiple Variable

In this part, you will implement linear regression with multiple variables to
predict the prices of houses. Suppose you are selling your house and you
want to know what a good market price would be. One way to do this is to
first collect information on recent houses sold and make a model of housing
prices.

The first column is the size of the house (in square feet), the
second column is the number of bedrooms, and the third column is the price
of the house.

```python
# Load the second dataset
data = np.loadtxt('machine-learning-ex1/ex1/ex1data2.txt', delimiter=",")
```

### 3.1 Feature Normalization

- Subtract the mean value of each feature from the dataset.
- After subtracting the mean, additionally scale (divide) the feature values by their respective "standard deviations."

```python
# Define the feature normalization function
def normalize_features(data):
    
    means = []
    stds = []
    norm_data = np.zeros(data.shape)
    
    # For each feature, store the mean, standard deviation, and normalized data
    for feature in range(data.shape[1]):
        means.append(np.mean(data[:,feature]))
        stds.append(np.std(data[:,feature]))
        norm_data[:,feature] = (data[:,feature] - np.mean(data[:,feature]))/np.std(data[:,feature])
        
    return norm_data, np.array(means), np.array(stds)
```

```python
# Normalize the data and store the means and standard deviations
norm_data, means, stds = normalize_features(data)
X_norm, y_norm = norm_data[:, :2], norm_data[:, 2]
```

```python
# Add a column of ones to X as X_0
X_norm_col = np.hstack((np.ones((X_norm.shape[0], 1)), X_norm))

# Change y into a column vector
y_norm_col = y_norm.reshape(-1,1)

# Initiliaze the parameter vector
theta_multi = np.zeros((1,3))
```

### 3.2 Selecting Learning Rates

```python
# Define a function to find the best learning rate

def select_learning_rate(alphas, X, y, theta, iterations):
    
    costs = []
    
    # For every learning rate, calculate the final cost function and store it
    for alpha in alphas:
        final_theta = gradient_descent(X, y, theta, alpha, iterations)[-1]
        costs.append(cost_function(X, y, final_theta))
        
    return costs
```

```python
# Set the number of iterations and alphas and find the final cost function for each alpha
iters = 50
alphas = [0.3, 0.1, 0.03, 0.01, 0.003, 0.001, 0.0003, 0.0001]
costs = select_learning_rate(alphas, X_norm_col, y_norm_col, theta_multi.T, iters)
```

```python
# Plot the cost function results against the alphas
plt.figure(figsize=(8,6))
plt.plot(alphas, costs, marker = 'x', color='r')
plt.title("Learning Rate vs Final Cost Function")
plt.xlabel("Learning Rate")
plt.ylabel("Cost Function Output")
plt.show()
```

```python
# Store the thetas from the gradient descent when alpha = 0.1
thetas = gradient_descent(X_norm_col, y_norm_col, theta_multi.T, 0.1, iters)
theta_costs = []

# Calculate the cost function at each theta vector
for theta in thetas:
    theta_costs.append(cost_function(X_norm_col, y_norm_col, theta))
    
# Plot the cost function results against the gradient descent
plt.figure(figsize=(8,6))
plt.plot(np.arange(0,51,1), theta_costs, color='b')
plt.title("Gradient Descent Iterations vs Final Cost Function (when alpha = 0.1)")
plt.xlabel("Iterations")
plt.ylabel("Cost Function Output")
plt.show()
```

### 3.3 Normal Equation

```python
# Define the normal equation to calculate the theta parameters

def normal_equation(X, y):
    
    step1 = np.linalg.inv(np.matmul(X.T, X))
    step2 = np.matmul(step1, X.T)
    theta = np.matmul(step2, y)
    
    return theta
```

```python
# Get X, y
X, y = data[:, :2], data[:, 2]

# Add a column of ones to X as X_0
X_col = np.hstack((np.ones((X.shape[0], 1)), X))

# Change y into a column vector
y_col = y.reshape(-1,1)

# Find the theta parameters
NE_theta = normal_equation(X_col, y_col)
print(NE_theta)
```

```python
# Calculate the predicted values
line = NE_theta[0] + X_col[:,1]*NE_theta[1] + X_col[:,2]*NE_theta[2]

# Create a line with slope 1 to show perfect correlation
x1 = np.linspace(1e5,7e5)
y1 = x1

# Plot the predicted vs actuals
plt.plot(x1, y1, 'r')
plt.scatter(line, y)
plt.title('Predicted vs Actuals Using Normal Equation')
plt.xlabel("Predicted")
plt.ylabel("Actuals")
plt.show()
```

Compare the normal equation results with the gradient descent results

```python
# Calculate the thetas using gradient descent
GD_thetas = gradient_descent(X_norm_col, y_norm_col, theta_multi.T, 0.3, 10000)
GD_theta = GD_thetas[-1]
print(GD_theta)
```

```python
# Calculate the predicted values
line = GD_theta[0] + X_norm_col[:,1]*GD_theta[1] + X_norm_col[:,2]*GD_theta[2]

# Create a line with slope 1 to show perfect correlation
x2 = np.linspace(-2.0,3)
y2 = x2

# Plot the predicted vs actuals
plt.plot(x2, y2, 'r')
plt.scatter(line, y_norm)
plt.title('Predicted vs Actuals Using Gradient Descent')
plt.xlabel("Predicted")
plt.ylabel("Actuals")
plt.show()
```
